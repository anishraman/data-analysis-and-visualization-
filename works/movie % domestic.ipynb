```python
    from matplotlib import pyplot as plt
    import seaborn as sns
    %matplotlib inline
    import warnings
    warnings.filterwarnings('ignore')
```

```python
sns.set_style("darkgrid")
f,axes=plt.subplots(2, 2, figsize=(15,15))
#plt 0,0
k1 =sns.kdeplot(movies["BudgetMillions"],movies["Audience Rating"],\
                shade=True,shade_lowest=True,cmap='inferno',ax=axes[0,0])
k1b =sns.kdeplot(movies["BudgetMillions"],movies["Audience Rating"],\
                cmap='cool',ax=axes[0,0])
#plt 0,1
k2 =sns.kdeplot(movies["BudgetMillions"],movies["Critic rating"],\
                shade=True,shade_lowest=True,cmap='inferno',ax=axes[0,1])
k2b =sns.kdeplot(movies["BudgetMillions"],movies["Critic rating"],\
                cmap='cool',ax=axes[0,1])
#plt 1,0
z = sns.violinplot(data=movies,x='Year',y='BudgetMillions',ax=axes[1,0],\
                  palette='YlOrRd')
#plt 1,1
k4 =sns.kdeplot(movies["Critic rating"],movies["Audience Rating"],shade=True,shade_lowest=False,cmap='Blues_r',ax=axes[1,1])

k4b =sns.kdeplot(movies["Critic rating"],movies["Audience Rating"],\
               cmap='gist_gray_r',ax=axes[1,1])
    
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
sns.set_style("darkgrid",{"axes.facecolor":"black"})
f,axes=plt.subplots(2, 2, figsize=(15,15))
#plt 0,0
k1 =sns.kdeplot(movies["BudgetMillions"],movies["Audience Rating"],\
                shade=True,shade_lowest=True,cmap='inferno',ax=axes[0,0])
k1b =sns.kdeplot(movies["BudgetMillions"],movies["Audience Rating"],\
                cmap='cool',ax=axes[0,0])
#plt 0,1
k2 =sns.kdeplot(movies["BudgetMillions"],movies["Critic rating"],\
                shade=True,shade_lowest=True,cmap='inferno',ax=axes[0,1])
k2b =sns.kdeplot(movies["BudgetMillions"],movies["Critic rating"],\
                cmap='cool',ax=axes[0,1])
#plt 1,0
z = sns.violinplot(data=movies,x='Year',y='BudgetMillions',ax=axes[1,0],\
                  palette='YlOrRd')
#plt 1,1
k4 =sns.kdeplot(movies["Critic rating"],movies["Audience Rating"],shade=True,shade_lowest=False,cmap='Blues_r',ax=axes[1,1])

k4b =sns.kdeplot(movies["Critic rating"],movies["Audience Rating"],\
               cmap='gist_gray_r',ax=axes[1,1])
    
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
list1 =list()
mylabels=list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre ==gen].BudgetMillions)
    mylabels.append(gen)
sns.set_style("whitegrid")
fig, ax =plt.subplots()
fig.set_size_inches(11.7,8.27)
h =plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels,color={"Red","salmon","Black","Pink","Brown","Green","Yellow"})
plt.title("Movie Budget distribution",fontsize=35,\
         color="DarkBlue",fontname="Times new roaman")
plt.ylabel("Number of movies",fontsize=25,color="black")
plt.xlabel("Budget",fontsize=25,color="black")
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(frameon=True, fancybox=True,shadow=True,framealpha=0.5,prop={'size':20})
plt.show()
```
